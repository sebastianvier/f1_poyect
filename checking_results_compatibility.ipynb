```python
import numpy as np
import pandas as pd
```

```python
## Get the data from csv

path = './data/'

grid = pd.read_csv(path + 'f1_grid.csv')
grid.columns = ['position', 'time', 'driver', 'constructor', 'race']
results = pd.read_csv(path + 'f1_results.csv')
results.columns = ['position', 'laps', 'time', 'points', 'driver', 'constructor', 'race']
```

```python
complete_merge = pd.merge(grid,results,how='outer', on=['driver', 'race'])
```

```python
race_additions = complete_merge[complete_merge['position_x'].isna()]
```

```python
grid_additions = complete_merge[(complete_merge['position_y'].isna()) & (complete_merge['laps'].isna())]
```
